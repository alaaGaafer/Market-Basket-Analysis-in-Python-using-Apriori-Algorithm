```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
import numpy as np
```

```python
dataset= pd.read_csv("/content/groceries - groceries.csv")
# df = pd.read_csv("../input/market-basket-analysis/Assignment-1_Data.csv", sep=';')
dataset
```

```python
print(dataset.info())
print(dataset.shape)
```

```python
dataset
```

```python
#seeing the number of nan values in each column to decide which columns to drop
for column in dataset:
  print(column,dataset[column].isnull().sum())
```

```python
#dropping all the columns the number of nans sxceeds 9000 in them and the items column so it doesn't show in the frequent items list
dataset = dataset.drop(dataset.iloc[:, 12:32],axis = 1)
dataset = dataset.drop(["Item(s)"],axis = 1)
dataset
```

```python
dataset.info()
```

```python
#making a list of list to contain each transaction except for the nans
transactions = []
for i in range(len(dataset)):
    row = []
    for j in range(5):
        if isinstance(dataset.iloc[i,j], str):
            row.append(str(dataset.iloc[i,j]))
    transactions.append(row)
transactions
```

```python
#Converting the transaction data to a one-hot encoded forma
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
frequent_itemsets = apriori(df, min_support=0.03, use_colnames=True)
frequent_itemsets.sort_values(by="support",ascending=False)
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.sort_values(by="support",ascending=False)
```
